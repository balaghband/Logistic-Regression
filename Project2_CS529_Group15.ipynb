```python

train_files = []
for i in range(10):
  train_files.append('HM2/metal.0000'+str(i)+'.au')

for j in range(11,90):
  train_files.append('HM2/metal.000'+str(j)+'.au')

train_files
```

Feature Extraction

```python
short1_fourier= []
mel2_frequency = []
chroma3_list = []
special4_contrast_list = []
zero_crossing_list = []
zero_crossing_rate_list = []
energy_list = []
audio_index = []
features =[]
labels = []
frame_size = 2048
hop_size = 512

for i in train_files:
    audio_index.append(i)
    audio_sample, sampling_rate = librosa.load(i)

    stft = librosa.stft(audio_sample)
    short1_fourier.append(stft)

    mfcc = librosa.feature.mfcc(y=audio_sample,
                                sr=sampling_rate)
    mel2_frequency.append(mfcc)

    chroma = librosa.feature.chroma_stft(y=audio_sample, sr=sampling_rate)
    chroma3_list.append(chroma)

    spectral_contrast = librosa.feature.spectral_contrast(y=audio_sample,
                                                          sr=sampling_rate)
    special4_contrast_list.append(spectral_contrast)

    zero_crossings = librosa.zero_crossings(audio_sample, pad=False)
    zero_crossing_list.append(zero_crossings)
    zero_crossing_rate = sum(zero_crossings) / len(audio_sample)
    zero_crossing_rate_list.append(zero_crossing_rate)
    energy = sum(abs(audio_sample)**2) / len(audio_sample)
    energy_list.append(energy)


audio_features = pd.DataFrame({'audio':audio_index,
                               'short':short1_fourier,
                               'mel':mel2_frequency,
                               'chrom':chroma3_list,
                               'special':special4_contrast_list,
                               'zero_c':zero_crossing_list,
                               'zero_rate':zero_crossing_rate_list,
                               'energy':energy_list})

audio_features
```
